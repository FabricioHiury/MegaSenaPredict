```python
import numpy as np
import pandas as pd
import random
```

```python
df = pd.read_csv('mega_sena-1.csv')
df.head()
```

```python
df_filtered = df[df['Valor_Acumulado'] == 0.0]
df_filtered.head()
```

```python
numbers = ['Dezena_1', 'Dezena_2', 'Dezena_3', 'Dezena_4', 'Dezena_5', 'Dezena_6']
```

```python
dfNum = df_filtered[numbers]
dfNum.head()
```

```python
df_melted = dfNum.melt(var_name='Column', value_name='Numbers')
df_melted['Column'] = df_melted['Column'].str.split('_').str[-1] 
```

```python
count_numbers = df_melted['Numbers'].value_counts()
percent_numbers = (count_numbers / len(df_melted)) * 100
dfPercent = pd.DataFrame({'Numbers': percent_numbers.index, 'Percent': percent_numbers.values})
```

```python
dfPercent = dfPercent.sort_values(by='Percent', ascending=False)
```

```python
top_20 = dfPercent.head(20)
print(top_20)
```

```python
numbers_top_20 = top_20['Numbers'].tolist()
random_numbers = random.sample(numbers_top_20, 6)
random_numbers_sorted = sorted(random_numbers)
```

```python
print('Numbers drawn: ', random_numbers_sorted)
```
